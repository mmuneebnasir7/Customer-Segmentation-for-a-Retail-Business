```python
# Load dataset
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
df = pd.read_csv(r'C:\Users\ASAA\Downloads\Mall_Customers.csv')

# Display first few rows
print("First 5 rows of the dataset:")
print(df.head())

# Select relevant features
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans with K=3
kmeans3 = KMeans(n_clusters=3, random_state=42)
df['Cluster_3'] = kmeans3.fit_predict(X_scaled)

# Visualize the clusters for K=3
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster_3', palette='Set2', s=100)
plt.title('Customer Segmentation (K=3)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
print("\nCluster Summary for K=3:")
print(df.groupby('Cluster_3')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean())
```

```python
# Create KMeans clustering for K=4
kmeans4 = KMeans(n_clusters=4, random_state=42)
df['Cluster_4'] = kmeans4.fit_predict(X_scaled)
# Visualize the clusters for K=4
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster_4', palette='Set2', s=100)
plt.title('Customer Segmentation (K=4)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
# Describe the clusters for K=4
print("\nCluster Summary for K=4:")
print(df.groupby('Cluster_4')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean())
```

```python
# Create KMeans clustering for K=5
kmeans5 = KMeans(n_clusters=5, random_state=42)
df['Cluster_5'] = kmeans5.fit_predict(X_scaled)

# Visualize the clusters for K=5
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster_5', palette='Set3', s=100)
plt.title('Customer Segmentation (K=5)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Describe the clusters for K=5
print("\nCluster Summary for K=5:")
print(df.groupby('Cluster_5')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean())
```

```python
# Calculate WCSS for k = 3 to 5
wcss = []
for k in range(3, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow graph for K = 3 to 5
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(range(3, 6), wcss, marker='o', linestyle='--')
plt.title('Elbow Method (K = 3 to 5)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()

```

```python
# Apply KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels to original dataframe
df['Cluster'] = clusters

# Display first few results
print(df.head())
```

```python
# Apply PCA for 2D visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)

# Add PCA results to dataframe
df['PCA1'] = pca_components[:, 0]
df['PCA2'] = pca_components[:, 1]

# Plot the clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set1', s=100)
plt.title('Customer Segments Visualized (PCA 2D)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Group by clusters and view statistics
cluster_summary = df.groupby('Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
print("Cluster Summary (Average Values):\n")
print(cluster_summary)

# Optional: Count of customers per cluster
print("\nNumber of Customers in Each Cluster:")
print(df['Cluster'].value_counts())
```
